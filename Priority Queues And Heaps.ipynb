```python
'''Max-Priority Queue using Max heap Property'''
class pqnode:
    def __init__(self,value,priority):
        self.value=value
        self.priority=priority
class priorityQueue:
    def __init__(self):
        self.pqlist=[]
        self.size=0
    def getmax(self):
        if not self.size:
            return None
        return self.pqlist[0].value
    def getsize(self):
        return self.size
    
    def insert(self,value,priority):
        self.pqlist.append(pqnode(value,priority))
        
        i=self.size
        while i>0:
            if self.pqlist[(i-1)//2].priority<priority:
                self.pqlist[(i-1)//2],self.pqlist[i]=self.pqlist[i],self.pqlist[(i-1)//2]
                i=(i-1)//2
            else:
                break
        self.size+=1
    def removemax(self):
        k=self.pqlist[0]
        self.pqlist[0]=self.pqlist[self.size-1]
        
        self.pqlist.pop()
        
        self.size-=1
        i=0
        
        
        while i<self.size:
            if 2*i+1>self.size-1:
                break
            elif 2*i+2>self.size-1:
                if self.pqlist[2*i+1].priority>self.pqlist[i].priority:
                    self.pqlist[i],self.pqlist[2*i+1]=self.pqlist[2*i+1],self.pqlist[i]
                    i=2*i+1 
                break    
            elif self.pqlist[i].priority>=self.pqlist[2*i+1].priority and self.pqlist[i].priority>=self.pqlist[2*i+2].priority:
                break
            elif self.pqlist[2*i+1].priority>self.pqlist[2*i+2].priority:
                self.pqlist[i],self.pqlist[2*i+1]=self.pqlist[2*i+1],self.pqlist[i]
                i=2*i+1
            else:
                self.pqlist[i],self.pqlist[2*i+2]=self.pqlist[2*i+2],self.pqlist[i]
                i=2*i+2
        return k.value
                
    def print_pq(self):
        l=[[i.value,i.priority] for i in self.pqlist]
        print(l)
    
p=priorityQueue()

l=[['a',5],['b',10],['c',23],['d',1],['e',6],['f',20],['g',15],['h',23],['i',100],['j',9]]
for i in range(10):
    p.insert(l[i][0],l[i][1])
p.print_pq()
print(p.removemax())
p.print_pq()
print(p.removemax())
p.print_pq()
```

```python
'''Heap -Sort Algorithm'''
class Minheap:
    def __init__(self):
        self.pqlist=[]
        self.size=0
    
    def insert(self,value):
        self.pqlist.append(value)
        i=self.size
        while i>0:
            if self.pqlist[(i-1)//2]>value:
                self.pqlist[(i-1)//2],self.pqlist[i]=self.pqlist[i],self.pqlist[(i-1)//2]
                i=(i-1)//2
            else:
                break
        self.size+=1
    def removemin(self):
        k=self.pqlist[0]
        self.pqlist[0]=self.pqlist[self.size-1]
        self.pqlist.pop()
        self.size-=1
        i=0
        while i<self.size:
            if 2*i+1>self.size-1:
                break
            elif 2*i+2>self.size-1:
                if self.pqlist[2*i+1]<self.pqlist[i]:
                    self.pqlist[i],self.pqlist[2*i+1]=self.pqlist[2*i+1],self.pqlist[i]
                    i=2*i+1 
                break    
            elif self.pqlist[i]<=self.pqlist[2*i+1] and self.pqlist[i]<=self.pqlist[2*i+2]:
                break
            elif self.pqlist[2*i+1]<self.pqlist[2*i+2]:
                self.pqlist[i],self.pqlist[2*i+1]=self.pqlist[2*i+1],self.pqlist[i]
                i=2*i+1
            else:
                self.pqlist[i],self.pqlist[2*i+2]=self.pqlist[2*i+2],self.pqlist[i]
                i=2*i+2
        return k
                
    
heap=Minheap()
l=[5,2,7,1,9,26,3,4,12,21,0]
'''Inserting elements into Min-Heap'''
for i in range(11):
    heap.insert(l[i])

for i in range(11):
    print(heap.removemin(),end=' ')
```

```python
'''Inplace HeapSort Algorithm'''

class Minheap:
    def __init__(self):
        self.pqlist=[]
        self.size=0
        self.curr=1
    
    def heapify(self):
        while self.curr<self.size:
            i=self.curr
            value=self.pqlist[i]
            while i>0:
                if self.pqlist[(i-1)//2]>value:
                    self.pqlist[(i-1)//2],self.pqlist[i]=self.pqlist[i],self.pqlist[(i-1)//2]
                    i=(i-1)//2
                else:
                    break
            self.curr+=1
            
    def removemin(self):
        k=self.pqlist[0]
        self.pqlist[0]=self.pqlist[self.size-1]
        
        self.pqlist.pop()
        
        self.size-=1
        i=0
        
        
        while i<self.size:
            if 2*i+1>self.size-1:
                break
            elif 2*i+2>self.size-1:
                if self.pqlist[2*i+1]<self.pqlist[i]:
                    self.pqlist[i],self.pqlist[2*i+1]=self.pqlist[2*i+1],self.pqlist[i]
                    i=2*i+1 
                break    
            elif self.pqlist[i]<=self.pqlist[2*i+1] and self.pqlist[i]<=self.pqlist[2*i+2]:
                break
            elif self.pqlist[2*i+1]<self.pqlist[2*i+2]:
                self.pqlist[i],self.pqlist[2*i+1]=self.pqlist[2*i+1],self.pqlist[i]
                i=2*i+1
            else:
                self.pqlist[i],self.pqlist[2*i+2]=self.pqlist[2*i+2],self.pqlist[i]
                i=2*i+2
        return k
                
inheap=Minheap()    
inheap.size=int(input())
inheap.pqlist=list(map(int,input().split()))
inheap.heapify()
print('Min-Heap is ',inheap.pqlist)
print('Sorted array is ')
for i in range(inheap.size):
    print(inheap.removemin(),end=' ')
```

```python
'''Finding k smallest numbers in a list using Max-Heapify'''
def kSmallest(lst, k):
    l=lst[:k]
    heapq._heapify_max(l)
    for i in range(k,n):
        if l[0]>lst[i]:
            l[0]=lst[i]
            heapq._heapify_max(l)
    return l
'''Finding k Largest numbers in a list using Min-Heapify'''
def kLargest(lst, k):
    l=lst[:k]
    heapq.heapify(l)
    for i in range(k,n):
        if l[0]<lst[i]:
            l[0]=lst[i]
            heapq.heapify(l)
    return l

n=int(input())
lst=list(map(int,input().split()))
k=int(input())
ans=kSmallest(lst, k)
print(*ans, sep=' ')
ans=kLargest(lst, k)
print(*ans, sep=' ')
```

```python
def checkMaxHeap(lst,n,i=0):
    if i>n-1:
        return True
    elif 2*i+1>n-1:
        return True
    elif 2*i+2>n-1 and lst[i]<lst[2*i+1]:
        return False
    elif 2*i+2>n-1:
        return True
    elif lst[2*i+1]>lst[i] or lst[2*i+2]>lst[i]:
        return False
    else:
        return checkMaxHeap(lst,n,2*i+1) and checkMaxHeap(lst,n,2*i+2)
n1=int(input())
print(checkMaxHeap(list(map(int,input().split())),n1))
n2=int(input())
print(checkMaxHeap(list(map(int,input().split())),n2))
```
